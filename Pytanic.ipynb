```python
# Now Analyse about Pclass
titanic_df = pd.read_csv(r'C:\Users\Binay\Desktop\Kaggle\Titanic\train.csv')
sns.factorplot(x = 'Pclass', y='Survived', hue = 'Sex', data = titanic_df)
# Here we can make three categories based on the points in the graph
# Creating new Feature newPclass
titanic_df = pd.read_csv(r'C:\Users\Binay\Desktop\Kaggle\Titanic\train.csv')
titanic_df['newPclass'] = 1
titanic_df['newPclass'][(titanic_df.Pclass.isin([1,2])) & (titanic_df.Sex == 'female')] = 3
titanic_df['newPclass'][((titanic_df.Pclass == 3) & (titanic_df.Sex == 'female')) | ((titanic_df.Pclass == 1) & (titanic_df.Sex == 'male'))] = 2 

# Check the values in each new class for reference __ for testing purpose
titanic_df[['Survived', 'newPclass']][titanic_df.newPclass.isin([1])]
```

```python
# When feature created store it to final_titanic_df
final_titanic_df = titanic_df['newPclass'].to_frame()
final_titanic_df['Sex'] = titanic_df.Sex.map({'male': 0, 'female':1})
final_titanic_df.head()
```

```python
#Create new feature sibsp+parch
titanic_df['nFamilyMembers'] = titanic_df['SibSp'] + titanic_df['Parch']
titanic_df.groupby(by=['nFamilyMembers']).Survived.sum()
#family with less than 3 members are having better chance of survival and that is somewhat related to Age
# So creating new column for age band
def agebandFun(x):
    if 0<x<14:
        return 'Very child'
    elif 14<= x < 30:
        return 'Young'
    elif 30<= x < 50:
        return 'Aged'
    else:
        return 'Old'
        
titanic_df['Ageband'] = titanic_df.Age.apply(agebandFun)
```

```python
#titanic_df.groupby(by=['Ageband']).Survived.sum()
sns.factorplot(x = 'Ageband', y='Survived',data = titanic_df)
```
